```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data'
# importing the data by using website url
```

```python
# this is to let pytyhon know that this dataset has no header
data = pd.read_csv(url, header=None)
data.head()
```

```python
# this is to rename the columns by using the set_axis function
data.set_axis(['Flow_ID','Date','Time','Count'],axis=1,inplace=True)
```

```python
# selecting data for the date (7/24/2005 and having flow id=7)
date_filter = data['Date']=='07/24/05' # this is to slice to get the row that has the date 7/25/2005
flow_filter = data['Flow_ID']==7  #this is to slice to get the flow id of 7 
filter_data = data.loc[date_filter & flow_filter,:]
filter_data.head(10)
```
